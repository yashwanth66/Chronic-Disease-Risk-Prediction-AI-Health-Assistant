```python
import pandas as pd

df_imputed = df_selected.copy()

# ─────────────────────────────────────────────────────────────
# 1. DROP ROWS — Low importance features (drop rows with NaN)
# Physical Activity, Alcohol Consumption, Urban/Rural
# Missing %: 0.3%, 9.6%, 3.2% — but low importance so drop rows
# ─────────────────────────────────────────────────────────────
drop_row_cols = ["Physical Activity", "Alcohol Consumption", "Urban/Rural"]

before = len(df_imputed)
df_imputed = df_imputed.dropna(subset=drop_row_cols)
after = len(df_imputed)
print(f"Rows dropped: {before - after} ({((before - after)/before)*100:.2f}%)")
print(f"Remaining rows: {after}")

# ─────────────────────────────────────────────────────────────
# 2. CAREFUL IMPUTATION — High importance features (≥5%)
# Age, Pneumonia Vaccination, Physical Health Status,
# Oral Health, Health Status
# All are categorical/ordinal → MODE imputation
# ─────────────────────────────────────────────────────────────
careful_cols = [
    "Pneumonia Vaccination",   # 16.1% missing — highest risk
    "Physical Health Status",  # 2.4% missing
    "Oral Health",             # 2.1% missing
    "Health Status",           # 0.3% missing
    # Age has 0% missing — no imputation needed
]

for col in careful_cols:
    mode_val = df_imputed[col].mode(dropna=True)[0]
    missing_before = df_imputed[col].isna().sum()
    df_imputed[col] = df_imputed[col].fillna(mode_val)
    print(f"[CAREFUL] {col}: filled {missing_before} NaNs with mode={mode_val}")

# ─────────────────────────────────────────────────────────────
# 3. SIMPLE IMPUTATION — Medium importance features (1–5%)
# Categorical/Ordinal → MODE
# Continuous → MEDIAN
# ─────────────────────────────────────────────────────────────

# Categorical medium-importance → Mode
simple_mode_cols = [
    "Child Count",          # 2.0% missing — ordinal categorical
    "Tobacco Use",          # 7.0% missing — ordinal categorical
    "CHD or Heart Disease", # 1.1% missing — binary categorical
    "Sex",                  # 0% missing — already complete
    "Mental Health",        # 1.8% missing — ordinal categorical
    "Asthma",               # 0.4% missing — binary categorical
    "State",                # 1.6% missing — nominal categorical
]

for col in simple_mode_cols:
    missing_before = df_imputed[col].isna().sum()
    if missing_before > 0:
        mode_val = df_imputed[col].mode(dropna=True)[0]
        df_imputed[col] = df_imputed[col].fillna(mode_val)
        print(f"[SIMPLE-MODE] {col}: filled {missing_before} NaNs with mode={mode_val}")

# Continuous medium-importance → Median (robust to outliers)
simple_median_cols = [
    "BMI",             # 9.4% missing — continuous, has outliers
    "Weight (kg)",     # 7.95% missing — continuous, has outliers
    "Height (inches)", # 6.96% missing — continuous, has outliers
]

for col in simple_median_cols:
    missing_before = df_imputed[col].isna().sum()
    if missing_before > 0:
        median_val = df_imputed[col].median()
        df_imputed[col] = df_imputed[col].fillna(median_val)
        print(f"[SIMPLE-MEDIAN] {col}: filled {missing_before} NaNs with median={median_val:.2f}")

# ─────────────────────────────────────────────────────────────
# 4. VERIFY — No missing values should remain
# ─────────────────────────────────────────────────────────────
print("\n── Verification ──")
remaining = df_imputed.isna().sum()
if remaining.sum() == 0:
    print("✅ No missing values remaining!")
else:
    print("⚠️ Still missing:")
    print(remaining[remaining > 0])

print(f"\nFinal dataset shape: {df_imputed.shape}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.metrics import (accuracy_score, classification_report, confusion_matrix, roc_auc_score, roc_curve)
from xgboost import XGBClassifier

# ─────────────────────────────────────────────────────────────
# 1. PREP DATASET
# ─────────────────────────────────────────────────────────────
cols_to_drop = ["Physical Activity", "Alcohol Consumption", "Urban/Rural"]

df_model = df_imputed.dropna(subset=['Arthritis']).copy()
df_model = df_model.drop(columns=cols_to_drop)

print(f"Dataset shape: {df_model.shape}")
print(f"Missing values: {df_model.isna().sum().sum()}")
print(f"\nTarget distribution:\n{df_model['Arthritis'].value_counts(normalize=True).round(3)}")

# ─────────────────────────────────────────────────────────────
# 2. FEATURES & TARGET
# ─────────────────────────────────────────────────────────────
target_col = 'Arthritis'
X = df_model.drop(columns=[target_col])
y = df_model[target_col]

# Encode if needed
le = LabelEncoder()
for col in X.select_dtypes(include=['object', 'category']).columns:
    X[col] = le.fit_transform(X[col].astype(str))

# XGBoost needs integer labels starting from 0
y_encoded = le.fit_transform(y.astype(str))  # 1,2 → 0,1
print(f"\nEncoded target classes: {np.unique(y_encoded)}")

# ─────────────────────────────────────────────────────────────
# 3. TRAIN/TEST SPLIT — Stratified to preserve class balance
# ─────────────────────────────────────────────────────────────
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded,
    test_size=0.2,
    random_state=42,
    stratify=y_encoded
)
print(f"\nTrain size: {X_train.shape[0]}")
print(f"Test size : {X_test.shape[0]}")

# ─────────────────────────────────────────────────────────────
# 4. MODELS
# ─────────────────────────────────────────────────────────────
rf = RandomForestClassifier(
    n_estimators=50,
    max_depth=10,
    max_features='sqrt',
    min_samples_leaf=50,
    class_weight='balanced',
    random_state=42,
    n_jobs=-1
)

xgb = XGBClassifier(
    n_estimators=100,
    max_depth=6,
    learning_rate=0.1,
    subsample=0.8,
    colsample_bytree=0.8,
    scale_pos_weight=(y_encoded == 0).sum() / (y_encoded == 1).sum(),  # handles imbalance
    eval_metric='logloss',
    random_state=42,
    n_jobs=-1
)

models = {
    'Random Forest': rf,
    'XGBoost': xgb
}

# ─────────────────────────────────────────────────────────────
# 5. TRAIN + EVALUATE
# ─────────────────────────────────────────────────────────────
results = {}

for name, model in models.items():
    print(f"\n{'='*50}")
    print(f"  {name}")
    print(f"{'='*50}")
    
    # Train
    model.fit(X_train, y_train)
    
    # Predict
    y_pred  = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1]
    
    # Metrics
    acc     = accuracy_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_proba)
    report  = classification_report(y_test, y_pred, target_names=['No Arthritis', 'Arthritis'])
    cm      = confusion_matrix(y_test, y_pred)
    
    results[name] = {
        'model'    : model,
        'y_pred'   : y_pred,
        'y_proba'  : y_proba,
        'accuracy' : acc,
        'roc_auc'  : roc_auc,
        'cm'       : cm,
        'report'   : report
    }
    
    print(f"\nAccuracy : {acc:.4f}")
    print(f"ROC-AUC  : {roc_auc:.4f}")
    print(f"\nClassification Report:\n{report}")

# ─────────────────────────────────────────────────────────────
# 6. CROSS VALIDATION (5-Fold Stratified)
# ─────────────────────────────────────────────────────────────
print(f"\n{'='*50}")
print("  5-Fold Stratified Cross Validation")
print(f"{'='*50}")

# Sample for CV speed on large dataset
sample_idx = np.random.choice(len(X), size=min(50000, len(X)), replace=False)
X_cv = X.iloc[sample_idx]
y_cv = y_encoded[sample_idx]

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

for name, model in models.items():
    cv_scores = cross_val_score(model, X_cv, y_cv, cv=cv, scoring='roc_auc', n_jobs=-1)
    print(f"\n{name}")
    print(f"  CV ROC-AUC scores : {cv_scores.round(4)}")
    print(f"  Mean  : {cv_scores.mean():.4f}")
    print(f"  Std   : {cv_scores.std():.4f}")
    results[name]['cv_scores'] = cv_scores

# ─────────────────────────────────────────────────────────────
# 7. CONFUSION MATRICES
# ─────────────────────────────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

for ax, (name, res) in zip(axes, results.items()):
    sns.heatmap(
        res['cm'], annot=True, fmt='d', cmap='Blues', ax=ax,
        xticklabels=['No Arthritis', 'Arthritis'],
        yticklabels=['No Arthritis', 'Arthritis']
    )
    ax.set_title(f'{name}\nAccuracy: {res["accuracy"]:.4f} | AUC: {res["roc_auc"]:.4f}')
    ax.set_xlabel('Predicted')
    ax.set_ylabel('Actual')

plt.suptitle('Confusion Matrices', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

# ─────────────────────────────────────────────────────────────
# 8. ROC CURVES
# ─────────────────────────────────────────────────────────────
plt.figure(figsize=(8, 6))

colors = ['#2563eb', '#dc2626']
for (name, res), color in zip(results.items(), colors):
    fpr, tpr, _ = roc_curve(y_test, res['y_proba'])
    plt.plot(fpr, tpr, label=f"{name} (AUC = {res['roc_auc']:.4f})", color=color, lw=2)

plt.plot([0, 1], [0, 1], 'k--', lw=1, label='Random Classifier')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve Comparison')
plt.legend(loc='lower right')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# ─────────────────────────────────────────────────────────────
# 9. FINAL SUMMARY TABLE
# ─────────────────────────────────────────────────────────────
print(f"\n{'='*50}")
print("  FINAL MODEL COMPARISON")
print(f"{'='*50}")
print(f"{'Metric':<20} {'Random Forest':>15} {'XGBoost':>15}")
print("-" * 52)
print(f"{'Accuracy':<20} {results['Random Forest']['accuracy']:>15.4f} {results['XGBoost']['accuracy']:>15.4f}")
print(f"{'ROC-AUC':<20} {results['Random Forest']['roc_auc']:>15.4f} {results['XGBoost']['roc_auc']:>15.4f}")
print(f"{'CV Mean AUC':<20} {results['Random Forest']['cv_scores'].mean():>15.4f} {results['XGBoost']['cv_scores'].mean():>15.4f}")
print(f"{'CV Std AUC':<20} {results['Random Forest']['cv_scores'].std():>15.4f} {results['XGBoost']['cv_scores'].std():>15.4f}")

winner = max(results, key=lambda x: results[x]['roc_auc'])
print(f"\n🏆 Better Model (by ROC-AUC): {winner}")
```
